Semantic chunking

```python
from langchain_experimental.text_splitter import SemanticChunker

text_splitter = SemanticChunker(embeddings=embeddings, breakpoint_threshold_type='percentile', breakpoint_threshold_amount=90)

def json_to_documents(json_data):
    """
    Converts a JSON obj containing legal articles  into langchain Document

    Args:
        json_data: data in json format

    Returns:
        a list of Document
    """
    documents = []
    for article in json_data:
        article_title = article["title"]
        for paragraph in article["paragraphs"]:
            context = paragraph["context"]

            # Split the context into smaller documents
            docs = text_splitter.create_documents(context)

            # Append metadata to each document
            for doc in docs:
                # Append title and paragraph-level metadata
                doc.metadata = {
                    "title": article_title,
                }
                doc.page_content = context
                documents.append(doc)

    return documents

documents = json_to_documents(data["data"])
```

```python
import subprocess
import time
import threading

# Start the ollama server in a new process
process = subprocess.Popen(['ollama', 'serve'], stdout=subprocess.PIPE, stderr=subprocess.PIPE)

# Function to print server output
def print_output(process):
    while True:
        output = process.stdout.readline()
        if output == b'' and process.poll() is not None:
            break
        if output:
            print(output.strip().decode('utf-8'))
        time.sleep(1)

# Start a thread to print server output
thread = threading.Thread(target=print_output, args=(process,))
thread.start()

print("Ollama server is running in the background")
```

Query rewriting

```python
#constant
retriever = vectorstore.as_retriever()
```

```python
from langchain.prompts import PromptTemplate
rewrite_llm = llm

query_rewrite_template = """
You are an AI assistant tasked with improving user queries to be more specific and clear, especially when the original query is vague or unclear.
When rewriting, or creating infer the user's intent, remove ambiguity, and make the query actionable based on the given context.
Do not overcomplicate or add irrelevant details. Keep the rewritten query clear and focused. Make sure the question is answeralbe.

Only respond with the rewritten query, in the same language as the original query.

Original query: {original_query}
Context: {context}

Rewritten query:
"""


query_rewrite_prompt = PromptTemplate(
    input_variables=["original_query", "context"],
    template=query_rewrite_template
)
re_write_chain = query_rewrite_prompt | rewrite_llm

def rewrite_query(original_query):
    """
    Rewrite the original query to improve quality.
    """
    retrieved_context = retriever.get_relevant_documents(original_query)
    context = retrieved_context[0].page_content
    response = re_write_chain.invoke({"original_query": original_query, "context": context})
    return response.content

original_query = "tương tác là gì và đối tượng nào là một phần của hiến pháp?"
rewritten_query = rewrite_query(original_query)
print("Original query:", original_query)
print("\nRewritten query:", rewritten_query)
```

Step-back prompting: genenrate broader queries, this is for beam search queries, because the query use beam search is better quality but lower diversity


```python
step_back_llm = llm

step_back_template = """
You are an AI assistant tasked with generating broader, more general queries to increase diversity and explore wider possibilities.
Based on the original query, create a step-back query that is more open-ended and general, while considering the given context. Ensure the step-back query allows for diverse responses or interpretations and answeralbe.

Only respond with the step-back query, in the same language as the original query.

Original query: {original_query}
Context: {context}

Step-back query:
"""


step_back_prompt = PromptTemplate(
    input_variables=["original_query"],
    template=step_back_template
)

step_back_chain = step_back_prompt | step_back_llm

def generate_step_back_query(original_query):
    """
    Generate a step-back query to enhance diveristy.

    """
    retrieved_context = retriever.get_relevant_documents(original_query)
    context = retrieved_context[0].page_content
    response = step_back_chain.invoke({"original_query": original_query, "context": context})
    response_content = response.content

    return response_content

original_query = "đối tượng áp dụng của luật thương mại là gì?"
step_back_query = generate_step_back_query(original_query)
print("\nOriginal query:", original_query)
print("\nStep-back query:", step_back_query)
```

```python
prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "You are an assistant for a question-answering task. "
        "Your task is to extract relevant information from the provided context and provide a concise, clear answer. "
        "Ensure your answer is brief and directly addresses the core of the question, short and enough information"
        "Just give the answer, do not disturb with no relevance information"
        "The answer should be in the same language as the input.\n"
        "Structure the response as follows:\n"
        "answer: [Your concise answer]\n"
        "Context: {context}"
    ),
    (
        "human",
        "Question: {question}"
    )
])
```

```python
qa_chain = RetrievalQA.from_chain_type(
        llm,
        retriever=db.as_retriever(),
        chain_type_kwargs={"prompt": prompt}
    )
```

```python
user_query = "đối tượng áp dụng của luật thương mại là gì?"
rewritten_query = rewrite_query(user_query)
step_back_query = generate_step_back_query(user_query)
output = qa_chain.invoke(step_back_query)
```

```python
def process_rag_output(output):
    """
    Convert the given result into the 'qas' format used in QA datasets.
    In case of not chosing expand the documents, we generate uuid for each question, append the relevant document and the answer along with question
    Otherwise, just expand the document by appending the new answer.
    Args:
        result (dict): The result containing 'query' and 'result' fields.

    Returns:
        list: A list of questions and answers in the 'qas' format .
    """


    question = output["query"]
    answer = output["result"].replace("answer: ", "")

    return (question, answer)

```

```python
from tqdm import tqdm
def answer_generation(data):
    new_data = {
        "version": "v1.0",
        "data": []
    }
    for document in tqdm(data, desc="Generating answer"):
        data_obj = {}
        title = document["title"]
        paragraphs = document["paragraphs"]

        data_obj["title"] = title
        data_obj["paragraphs"] = []

        for paragraph in paragraphs:
            context = paragraph["context"]

            qas = paragraph["qas"]
            new_qas = []
            id1 = qas[0]["id"]
            top_k_query = qas[0]["question"]
            id2 = qas[1]["id"]
            beam_query = qas[1]["question"]

            rewritten_query = rewrite_query(top_k_query)
            step_back_query = generate_step_back_query(beam_query)
            top_k_output = process_rag_output(qa_chain.invoke(rewritten_query)) #(query, answer)
            beam_output = process_rag_output(qa_chain.invoke(step_back_query)) #(query, answer)

            new_qas.append({
                "question": top_k_output[0],
                "id": id1,
                "answers": [
                    {
                        "text": top_k_output[1]
                    }
                ]
            })

            new_qas.append({
                "question": beam_output[0],
                "id": id2,
                "answers": [
                    {
                        "text": beam_output[1]
                    }
                ]
            })

            paragraph_obj = {
                "context": context,
                "qas": new_qas
            }
            data_obj["paragraphs"].append(paragraph_obj)

        new_data["data"].append(data_obj)

    return new_data
```
